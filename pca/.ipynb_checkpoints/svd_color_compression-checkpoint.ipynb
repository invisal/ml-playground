```python
import matplotlib.pyplot as plt
import matplotlib.image as pimg
import numpy as np
```

```python
img = pimg.imread('nature.jpg')
plt.imshow(img)
```

```python
kimg = np.zeros(img.shape, np.int32)
k = 200

for d in range(3):
  u, s, vt = np.linalg.svd(img[:,:,d])
  ku = u[:,:k]
  ks = np.diag(s[:k])
  kvt = vt[:k]
  kimg[:,:,d] = np.dot(np.dot(ku,  ks), kvt)

kimg[kimg < 0] = 0
kimg[kimg > 255] = 255
plt.imshow(kimg)

original_size = img.shape[0] * img.shape[1]* img.shape[2]
compressed_size = (img.shape[0] * k + k + img.shape[1] * k) * img.shape[2]
saved = (1 - compressed_size / original_size) * 100
print("Original image size: {0} bytes".format(original_size))
print("Compressed image size: {0} bytes".format(compressed_size))
print("Saved {0}%".format(saved))
```
