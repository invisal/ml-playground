```python
import matplotlib.pyplot as plt
import matplotlib.image as pimg
import numpy as np
```

```python
img = pimg.imread('cat.jpg')
print(img)
print(img.shape)
```

```python
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```

```python
u, s, vt = np.linalg.svd(img)
print(u.shape, s.shape, vt.shape)
```

```python
k = 50
ku = u[:,:k]
ks = np.diag(s[:k])
kvt = vt[:50]
kimg = np.dot(np.dot(ku,  ks), kvt)

plt.imshow(kimg, cmap='gray', vmin=0, vmax=255)
```

```python
original_size = img.shape[0] * img.shape[1]
compressed_size = img.shape[0] * k + k + img.shape[1] * k
saved = (1 - compressed_size / original_size) * 100
print("Original image size: {0} bytes".format(original_size))
print("Compressed image size: {0} bytes".format(compressed_size))
print("Saved {0}%".format(saved))
```
